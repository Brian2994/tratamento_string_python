# **1° Vamos fazer o tratamento com variáveis Dummys**

Quando estamos fazendo o tratamento dos dados para criação de um modelo de Machine Learning ou IA precisamos deixar todos os dados como números.

Para este processo utilizamos variáveis Dummys

```python
# Convertir campos categoricos para variáveis Dummys
df_dados = pd.get_dummies(df_dados)
```

```python
# Verificar as 10 linhas iniciais do DataFrame
df_dados.head(10)
```

```python
# Verificando o novo shape dos dados
df_dados.shape
```

# **2° Vamos fazer o tratamento utilizando o OneHotEncoding do Scikit Learn**

```python
# Carregar o arquivo original novamente
df_dados = pd.read_excel(r'C:/Users/Brian/Documents/Visual Studio Code/data engineer/tratamento_string_python/dados.xlsx')
```

```python
# Verificamos o shape dos dados
df_dados.shape
```

```python
# Verificamos o tipo de dados
df_dados.info()
```

```python
# Vamos importar o pacote do Scikit Learn para fazer o tratamento dos dados
from sklearn.preprocessing import LabelEncoder
```

```python
# Cria o encoder e aplica OneHotEncoder
lb = LabelEncoder()

for var in ['PAIS','TRANSMISSAO','MODELO']:
    df_dados[var] = lb.fit_transform(df_dados[var])
```
